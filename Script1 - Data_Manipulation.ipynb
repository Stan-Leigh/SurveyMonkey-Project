```python
import pandas as pd
import os
```

```python
pwd = os.getcwd() # pwd -> present working directory
```

```python
# Read the data

df = pd.read_excel(pwd + '\\Data - Survey Monkey Output Edited.xlsx', sheet_name='Edited_Data')
df.head()
```

```python
# Dimension of the data

df.shape
```

```python
df_modified = df.copy() # Create a copy to be able to rollback changes
```

```python
df_modified.columns
```

```python
columns_to_drop = ['Start Date', 'End Date', 'Email Address', 'First Name', 'Last Name', 'Custom Data 1']

df_modified.drop(columns=columns_to_drop, inplace=True)

df_modified.head()
```

```python
# Melt the data into the required format

id_vars = list(df_modified.columns)[:8]
value_vars = list(df_modified.columns)[8:]

df_melted = df_modified.melt(id_vars=id_vars, value_vars=value_vars, var_name='Question + SubQuestion', value_name='Answer')
df_melted.head(10)
```

```python
df1 = pd.read_excel(pwd + '\\Data - Survey Monkey Output Edited.xlsx', sheet_name='Question')
df1.head()
```

```python
df1_modified = df1.copy()
df1_modified.drop(columns=['Raw Question', 'Subquestion', 'Raw Subquestion'], inplace=True)
```

```python
df1_modified.head(9)
```

```python
# Merge the two dataframes to find the original question for every question_and_subquestion

df_merged = df_melted.merge(df1_modified, how='left', left_on='Question + SubQuestion', right_on='Question + Subquestion')

# Check if the merge happened correctly
print(f'Original data: {len(df_melted)} rows')
print(f'Merged data: {len(df_merged)} rows')

df_merged.head()
```

```python
# How many people answered each question?

respondents = df_merged[df_merged['Answer'].notna()]
respondents = respondents.groupby('Question')['Respondent ID'].nunique().reset_index()
respondents.rename(columns={'Respondent ID': 'Respondents'}, inplace=True)
respondents
```

```python
# Merge the two dataframes to find the number of respondents that answered each question

df_merged_two = df_merged.merge(respondents, how='left', left_on='Question', right_on='Question')

# Check if the merge happened correctly
print(f'Original data: {len(df_merged)} rows')
print(f'Merged data: {len(df_merged_two)} rows')

df_merged_two.head()
```

```python
# How many people gave the same answer for each question?

same_answer = df_merged.copy() # [df_merged['Answer'].notna()]
same_answer = same_answer.groupby(['Question + Subquestion', 'Answer'])['Respondent ID'].nunique().reset_index()
same_answer.rename(columns={'Respondent ID': 'Same Answer'}, inplace=True)
same_answer
```

```python
df_merged_three = df_merged_two.merge(same_answer, how='left', left_on=['Question + Subquestion', 'Answer'], right_on=['Question + Subquestion', 'Answer'])
df_merged_three['Same Answer'].fillna(0, inplace=True)

# Check if the merge happened correctly
print(f'Original data: {len(df_merged_two)} rows')
print(f'Merged data: {len(df_merged_three)} rows')

df_merged_three.head()
```

```python
output = df_merged_three.copy()
output.rename(columns={'Identify which division you work in. - Response': 'Division Primary',
                        'Identify which division you work in. - Other (please specify)': 'Division Secondary',
                        'Which of the following best describes your position level? - Response': 'Position',
                        'Which generation are you apart of? - Response': 'Generation',
                        'Please select the gender in which you identify. - Response': 'Gender',
                        'Which duration range best aligns with your tenure at your company? - Response': 'Tenure',
                        'Which of the following best describes your employment type? - Response': 'Employment type'}, 
                        inplace=True)
output.head()
```

```python
output.to_excel(pwd + '\\Final_Output.xlsx', index=False)
```
